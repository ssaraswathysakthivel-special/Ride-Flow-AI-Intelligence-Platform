```python
from groq import Groq
import os
import joblib

os.environ["GROQ_API_KEY"] = "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"

client = Groq()
```

```python
system_prompt = """
You are RideFlow AI Assistant.
Respond in English, Hindi, or Tamil based on user language.
Be short and helpful.
Ride Info:
- Driver: Arun
- ETA: 5 minutes
- Fare: ₹240
"""
```

```python
def rideflow_chatbot(user_input):

    completion = client.chat.completions.create(
        model="openai/gpt-oss-120b",
        messages=[
            {"role": "system", "content": system_prompt},
            {"role": "user", "content": user_input}
        ],
        temperature=0.3,
        max_completion_tokens=300,
        stream=False
    )

    return completion.choices[0].message.content
```

# Interactive command-line chatbot loop

```python
print("🚗 RideFlow Chatbot Started (type 'exit' to stop)\n")

while True:
    user_input = input("You: ")

    if user_input.lower() == "exit":
        print("Bot: Bye! Safe rides 🚗")
        break

    reply = rideflow_chatbot(user_input)
    print("Bot:", reply)
```
